```python
# ============================================================
#  U-Net  |  DeepFashion2 Top-5  |  TRAINING FROM SCRATCH
#  Encoder: ResNet-34 randomly initialised (no pretrained weights)
#  Decoder: Symmetric U-Net with skip connections
#  Post-processing: Connected components → BBoxes → Majority vote
#  Metrics: mAP, mIoU, Dice, P/R/F1, ROC/AUC  (pandas output)
# ============================================================

import os, json, time, warnings
import cv2
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.models as tv_models
import matplotlib.pyplot as plt
import pandas as pd
from torch.utils.data import Dataset, DataLoader
from torchvision.ops import box_iou
from torchmetrics.detection.mean_ap import MeanAveragePrecision
from sklearn.metrics import roc_curve, auc
from scipy import ndimage as ndi
from tqdm import tqdm
import logging

import sys
warnings.filterwarnings("ignore")

# Force-flush logging so Kaggle shows output immediately instead of buffering
class FlushHandler(logging.StreamHandler):
    def emit(self, record):
        super().emit(record)
        self.flush()

handler = FlushHandler(sys.stdout)
handler.setFormatter(logging.Formatter(
    "%(asctime)s  [%(levelname)-8s]  %(message)s", datefmt="%H:%M:%S"
))
logging.basicConfig(level=logging.INFO, handlers=[handler], force=True)
log = logging.getLogger("unet_scratch")

# cudnn auto-tunes kernels for fixed input size → ~10-15% faster
torch.backends.cudnn.benchmark = True

# ── Paths ─────────────────────────────────────────────────────
TRAIN_JSON = "/kaggle/input/datasets/hades1998/subsampling-json/subsampled_train_data.json"
TRAIN_DIR  = "/kaggle/input/datasets/hrushi1998/vr-mini-proj-dataset/Trimmed_Dataset/trimmed_train_data"
VAL_DIR    = "/kaggle/input/datasets/hrushi1998/vr-mini-proj-dataset/Trimmed_Dataset/trimmed_val_data"
VAL_JSON   = os.path.join(VAL_DIR, "processed_val_data.json")
SAVE_PATH  = "/kaggle/working/unet_scratch_best.pth"
ROC_PATH   = "/kaggle/working/unet_scratch_roc.png"

# ── Class mapping ─────────────────────────────────────────────
# pixel 0 = background
# pixel N = foreground class N  (N = 1..5)
TOP5        = [1, 8, 7, 2, 9]              # DF2 category IDs
LABEL_MAP   = {c: i + 1 for i, c in enumerate(TOP5)}   # cat_id → pixel value
CLASS_NAMES = {
    1: "Short Sleeve Top",
    2: "Trousers",
    3: "Shorts",
    4: "Long Sleeve Top",
    5: "Skirt",
}
NUM_CLASSES = 6    # 0=background + 5 foreground

# ── Hyperparameters ───────────────────────────────────────────
INPUT_SIZE    = 320     # resize all images to this on CPU (saves GPU transfer)
BATCH_SIZE    = 16      # U-Net is lighter than Mask R-CNN — can go higher
STAGE1_EPOCHS = 0       # no frozen stage — encoder has no pretrained weights to protect
STAGE2_EPOCHS = 20      # train end-to-end from random init — needs more epochs than transfer
STAGE1_LR     = 3e-4    # unused (STAGE1_EPOCHS=0) — kept for reference
STAGE2_LR     = 3e-4    # higher than transfer LR — random init needs stronger gradient signal
WEIGHT_DECAY  = 1e-4
GRAD_CLIP     = 1.0
SCORE_THRESH  = 0.5
IOU_THRESH    = 0.5
MIN_COMPONENT = 200     # min pixel area for a connected component to be kept

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
log.info("Device: %s  |  INPUT_SIZE: %d  |  BATCH: %d", DEVICE, INPUT_SIZE, BATCH_SIZE)


# ════════════════════════════════════════════════════════════
#  DATASET
#  Outputs: (image_tensor [3,H,W], seg_mask [H,W])
#  seg_mask pixel values: 0=background, 1..5=class
# ════════════════════════════════════════════════════════════
class FashionSegDataset(Dataset):
    def __init__(self, json_file: str, img_dir: str):
        self.img_dir = img_dir
        with open(json_file) as f:
            all_data = json.load(f)["data"]

        # Pre-filter missing images upfront so __getitem__ never returns None
        self.data = [
            rec for rec in all_data
            if os.path.exists(os.path.join(img_dir, rec["image_path"]))
        ]
        log.info("Dataset: %d valid records (from %d total)",
                 len(self.data), len(all_data))

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item     = self.data[idx]
        img_path = os.path.join(self.img_dir, item["image_path"])

        # ── Fast load + resize on CPU worker ─────────────────
        bgr = cv2.imread(img_path)
        if bgr is None:
            return None
        real_H, real_W = bgr.shape[:2]
        bgr = cv2.resize(bgr, (INPUT_SIZE, INPUT_SIZE),
                         interpolation=cv2.INTER_LINEAR)
        rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

        scale_x = INPUT_SIZE / real_W
        scale_y = INPUT_SIZE / real_H

        categories = item.get("item_categories",       [])
        polygons   = item.get("segmentation_polygons", [])

        # ── Build semantic segmentation mask ─────────────────
        # pixel 0 = background; pixel N = class N (1-based).
        # Later classes overwrite earlier ones where polygons overlap
        # (consistent with standard semantic seg convention).
        seg_mask = np.zeros((INPUT_SIZE, INPUT_SIZE), dtype=np.int32)  # cv2.fillPoly requires int32 or uint8

        for i, cat_id in enumerate(categories):
            if cat_id not in LABEL_MAP:
                continue
            pixel_val = LABEL_MAP[cat_id]
            for ring in polygons[i]:
                if len(ring) < 6:
                    continue
                pts = np.array(ring, dtype=np.float32).reshape(-1, 2)
                pts[:, 0] = np.clip(pts[:, 0] * scale_x, 0, INPUT_SIZE - 1)
                pts[:, 1] = np.clip(pts[:, 1] * scale_y, 0, INPUT_SIZE - 1)
                cv2.fillPoly(seg_mask, [pts.astype(np.int32)], color=pixel_val)

        # Image → float32 tensor [0,1]
        img_tensor  = torch.from_numpy(rgb).permute(2, 0, 1).float() / 255.0
        mask_tensor = torch.from_numpy(seg_mask).long()   # (H, W) int64

        return img_tensor, mask_tensor


def collate(batch):
    batch = [b for b in batch if b is not None]
    if not batch:
        return ()
    imgs, masks = zip(*batch)
    return torch.stack(imgs), torch.stack(masks)


train_loader = DataLoader(
    FashionSegDataset(TRAIN_JSON, TRAIN_DIR),
    batch_size=BATCH_SIZE, shuffle=True,
    num_workers=4, pin_memory=True,
    persistent_workers=True, prefetch_factor=2,
    collate_fn=collate,
)
val_loader = DataLoader(
    FashionSegDataset(VAL_JSON, VAL_DIR),
    batch_size=BATCH_SIZE, shuffle=False,
    num_workers=4, pin_memory=True,
    persistent_workers=True, prefetch_factor=2,
    collate_fn=collate,
)
log.info("Train batches: %d  |  Val batches: %d",
         len(train_loader), len(val_loader))


# ════════════════════════════════════════════════════════════
#  U-NET ARCHITECTURE
#  Encoder : ResNet-34 (pretrained on ImageNet)
#            layer0 → 64ch  (stride 2)
#            layer1 → 64ch  (stride 1, after maxpool)
#            layer2 → 128ch (stride 2)
#            layer3 → 256ch (stride 2)
#            layer4 → 512ch (stride 2)
#  Bottleneck: 512ch
#  Decoder : 4 upsampling blocks with skip connections
#            each block: upsample → concat skip → conv×2
#  Head    : 1×1 conv → NUM_CLASSES logits
# ════════════════════════════════════════════════════════════
class ConvBlock(nn.Module):
    """Two consecutive Conv-BN-ReLU layers."""
    def __init__(self, in_ch: int, out_ch: int):
        super().__init__()
        self.block = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, padding=1, bias=False),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, 3, padding=1, bias=False),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True),
        )

    def forward(self, x):
        return self.block(x)


class UpBlock(nn.Module):
    """
    Bilinear upsample → concatenate skip → ConvBlock.
    in_ch  = channels from the previous decoder stage
    skip_ch = channels from the matching encoder skip
    out_ch  = output channels of this decoder stage
    """
    def __init__(self, in_ch: int, skip_ch: int, out_ch: int):
        super().__init__()
        self.up   = nn.Upsample(scale_factor=2, mode="bilinear", align_corners=False)
        self.conv = ConvBlock(in_ch + skip_ch, out_ch)

    def forward(self, x, skip):
        x = self.up(x)
        # Handle odd spatial dimensions from encoder
        if x.shape != skip.shape:
            x = nn.functional.interpolate(
                x, size=skip.shape[2:], mode="bilinear", align_corners=False
            )
        return self.conv(torch.cat([x, skip], dim=1))


class UNetResNet34(nn.Module):
    """
    U-Net with pretrained ResNet-34 encoder.

    Encoder channel map (ResNet-34):
      stem   : 64  ch   (after conv1 + bn1 + relu)
      layer1 : 64  ch
      layer2 : 128 ch
      layer3 : 256 ch
      layer4 : 512 ch   ← bottleneck

    Decoder mirrors this in reverse.
    """
    def __init__(self, num_classes: int, pretrained: bool = True):
        super().__init__()

        weights = tv_models.ResNet34_Weights.DEFAULT if pretrained else None
        enc     = tv_models.resnet34(weights=weights)

        # ── Encoder stages (saved as ModuleLists for easy freezing) ──
        self.enc_stem   = nn.Sequential(enc.conv1, enc.bn1, enc.relu)  # → 64ch, /2
        self.enc_pool   = enc.maxpool                                    # → 64ch, /4
        self.enc_layer1 = enc.layer1    # → 64ch,  /4
        self.enc_layer2 = enc.layer2    # → 128ch, /8
        self.enc_layer3 = enc.layer3    # → 256ch, /16
        self.enc_layer4 = enc.layer4    # → 512ch, /32  ← bottleneck

        # ── Decoder ───────────────────────────────────────────────────
        self.dec4 = UpBlock(512, 256, 256)   # up from /32 → /16, skip=layer3
        self.dec3 = UpBlock(256, 128, 128)   # up from /16 → /8,  skip=layer2
        self.dec2 = UpBlock(128, 64,  64)    # up from /8  → /4,  skip=layer1
        self.dec1 = UpBlock(64,  64,  64)    # up from /4  → /2,  skip=stem
        # Final upsample back to full resolution (/2 → /1)
        self.final_up   = nn.Upsample(scale_factor=2, mode="bilinear", align_corners=False)
        self.final_conv = nn.Sequential(
            ConvBlock(64, 32),
            nn.Conv2d(32, num_classes, kernel_size=1),
        )

    def forward(self, x):
        # ── Encoder ──────────────────────────────────────────
        s1 = self.enc_stem(x)           # [B, 64,  H/2,  W/2]
        s2 = self.enc_layer1(self.enc_pool(s1))  # [B, 64,  H/4,  W/4]
        s3 = self.enc_layer2(s2)        # [B, 128, H/8,  W/8]
        s4 = self.enc_layer3(s3)        # [B, 256, H/16, W/16]
        bottleneck = self.enc_layer4(s4) # [B, 512, H/32, W/32]

        # ── Decoder ──────────────────────────────────────────
        d4 = self.dec4(bottleneck, s4)  # [B, 256, H/16, W/16]
        d3 = self.dec3(d4, s3)          # [B, 128, H/8,  W/8]
        d2 = self.dec2(d3, s2)          # [B, 64,  H/4,  W/4]
        d1 = self.dec1(d2, s1)          # [B, 64,  H/2,  W/2]

        out = self.final_up(d1)         # [B, 64,  H,    W]
        return self.final_conv(out)     # [B, num_classes, H, W]


def build_model(pretrained: bool = True) -> nn.Module:
    model = UNetResNet34(num_classes=NUM_CLASSES, pretrained=pretrained)
    return model.to(DEVICE)


# ── Encoder layer groups — used for staged freezing ──────────
def _encoder_params(model: nn.Module):
    return list(model.enc_stem.parameters())   + \
           list(model.enc_pool.parameters())   + \
           list(model.enc_layer1.parameters()) + \
           list(model.enc_layer2.parameters()) + \
           list(model.enc_layer3.parameters()) + \
           list(model.enc_layer4.parameters())


def freeze_encoder(model: nn.Module) -> None:
    for p in _encoder_params(model):
        p.requires_grad = False
    trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
    log.info("Encoder frozen — trainable params: %s", f"{trainable:,}")


def unfreeze_all(model: nn.Module) -> None:
    for p in model.parameters():
        p.requires_grad = True
    trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
    log.info("All layers unfrozen — trainable params: %s", f"{trainable:,}")


# ════════════════════════════════════════════════════════════
#  LOSS
#  Combined CrossEntropy (pixel classification) +
#  Dice loss (handles class imbalance — background dominates).
# ════════════════════════════════════════════════════════════
class DiceLoss(nn.Module):
    def __init__(self, num_classes: int, smooth: float = 1.0):
        super().__init__()
        self.num_classes = num_classes
        self.smooth      = smooth

    def forward(self, logits: torch.Tensor, targets: torch.Tensor) -> torch.Tensor:
        probs   = torch.softmax(logits, dim=1)   # (B, C, H, W)
        # One-hot encode targets: (B, H, W) → (B, C, H, W)
        onehot  = torch.zeros_like(probs).scatter_(
            1, targets.unsqueeze(1), 1
        )
        # Compute dice per class, average across foreground only (skip class 0)
        dice_sum = 0.0
        for c in range(1, self.num_classes):
            p = probs[:, c].reshape(-1)
            g = onehot[:, c].reshape(-1)
            dice_sum += (2 * (p * g).sum() + self.smooth) / \
                        (p.sum() + g.sum() + self.smooth)
        return 1.0 - dice_sum / (self.num_classes - 1)


class SegLoss(nn.Module):
    """
    CE + Dice weighted combination.
    CE handles per-pixel accuracy; Dice handles class imbalance.
    """
    def __init__(self, num_classes: int, ce_weight: float = 0.5):
        super().__init__()
        self.ce        = nn.CrossEntropyLoss()
        self.dice      = DiceLoss(num_classes)
        self.ce_weight = ce_weight

    def forward(self, logits, targets):
        return (self.ce_weight * self.ce(logits, targets) +
                (1 - self.ce_weight) * self.dice(logits, targets))


# ════════════════════════════════════════════════════════════
#  POST-PROCESSING
#  Semantic mask → instance detections
#  Steps per spec:
#    a. Connected component analysis per class
#    b. Bounding box from each component
#    c. Instance label = class of that component (majority vote
#       is trivially the class itself since masks are per-class)
# ════════════════════════════════════════════════════════════
def postprocess_to_instances(seg_map: np.ndarray, min_area: int = MIN_COMPONENT):
    """
    Convert a (H, W) semantic segmentation map to instance detections.

    Parameters
    ----------
    seg_map  : np.ndarray (H, W) int  — pixel values 0..NUM_CLASSES-1
    min_area : int — minimum pixel area to keep a component

    Returns
    -------
    boxes  : list of [x1, y1, x2, y2]   (pixel coords)
    labels : list of int                 (1-indexed class)
    scores : list of float               (1.0 — U-Net has no confidence score;
                                          we use the foreground pixel fraction
                                          of the component as a proxy score)
    masks  : list of np.ndarray (H, W) bool  — binary mask per instance
    """
    boxes, labels, scores, masks = [], [], [], []

    for cls_idx in range(1, NUM_CLASSES):
        binary = (seg_map == cls_idx).astype(np.uint8)
        if binary.sum() == 0:
            continue

        # Connected component analysis (step a)
        num_labels, cc_map = cv2.connectedComponents(binary, connectivity=8)

        for comp_id in range(1, num_labels):   # 0 = background in cc_map
            comp_mask = (cc_map == comp_id)
            area      = comp_mask.sum()
            if area < min_area:
                continue

            # Bounding box (step b)
            rows = np.where(comp_mask.any(axis=1))[0]
            cols = np.where(comp_mask.any(axis=0))[0]
            x1, y1, x2, y2 = int(cols[0]), int(rows[0]), int(cols[-1]), int(rows[-1])
            if x2 <= x1 or y2 <= y1:
                continue

            # Instance label = class of this component (step c)
            # Majority vote is trivially cls_idx since the binary mask
            # was extracted per-class. Still correct per the spec.
            boxes.append([x1, y1, x2, y2])
            labels.append(cls_idx)
            # Proxy confidence score: fraction of class pixels in the component
            # relative to total class pixels in the image — larger components
            # get higher scores, which makes ROC meaningful.
            total_cls_pixels = binary.sum()
            scores.append(float(area) / (total_cls_pixels + 1e-6))
            masks.append(comp_mask)

    return boxes, labels, scores, masks


# ════════════════════════════════════════════════════════════
#  TRAINING
# ════════════════════════════════════════════════════════════
def run_epoch(model, loader, optimizer, scaler, criterion,
              epoch_label: str) -> float:
    model.train()
    total_loss = 0.0
    print(f"\n>>> {epoch_label} started — {len(loader)} batches", flush=True)
    bar = tqdm(loader, desc=epoch_label, leave=True, dynamic_ncols=True)

    for batch in bar:
        if not batch:
            continue
        imgs, masks = batch
        imgs  = imgs.to(DEVICE, non_blocking=True)
        masks = masks.to(DEVICE, non_blocking=True)

        optimizer.zero_grad(set_to_none=True)

        with torch.amp.autocast("cuda"):
            logits = model(imgs)                    # (B, C, H, W)
            loss   = criterion(logits, masks)

        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), GRAD_CLIP)
        scaler.step(optimizer)
        scaler.update()

        total_loss += loss.item()
        bar.set_postfix(loss=f"{loss.item():.4f}",
                        lr=f"{optimizer.param_groups[0]['lr']:.1e}")

    return total_loss / max(len(loader), 1)


def run_val_loss(model, loader, criterion) -> float:
    print("  Computing val loss...", flush=True)
    model.eval()
    total = 0.0
    with torch.no_grad():
        for batch in loader:
            if not batch:
                continue
            imgs, masks = batch
            imgs  = imgs.to(DEVICE, non_blocking=True)
            masks = masks.to(DEVICE, non_blocking=True)
            with torch.amp.autocast("cuda"):
                logits = model(imgs)
                total += criterion(logits, masks).item()
    return total / max(len(loader), 1)


def train_stage(model, train_loader, val_loader,
                criterion, epochs: int, lr: float,
                stage_name: str, best_val_loss: float,
                save_path: str) -> float:

    params    = [p for p in model.parameters() if p.requires_grad]
    optimizer = optim.AdamW(params, lr=lr, weight_decay=WEIGHT_DECAY)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(
        optimizer, T_max=epochs, eta_min=lr / 20
    )
    scaler = torch.amp.GradScaler("cuda")

    print(f"\n{'='*60}", flush=True)
    print(f"  {stage_name}  |  epochs={epochs}  lr={lr:.0e}", flush=True)
    print(f"{'='*60}", flush=True)
    log.info("=== %s  |  epochs=%d  lr=%.0e ===", stage_name, epochs, lr)
    rows = []

    for ep in range(1, epochs + 1):
        t0         = time.time()
        train_loss = run_epoch(model, train_loader, optimizer, scaler,
                               criterion, f"{stage_name} Ep{ep}/{epochs}")
        val_loss   = run_val_loss(model, val_loader, criterion)
        scheduler.step()
        elapsed    = time.time() - t0

        rows.append({
            "stage"      : stage_name,
            "epoch"      : ep,
            "train_loss" : round(train_loss, 4),
            "val_loss"   : round(val_loss,   4),
            "lr"         : f"{optimizer.param_groups[0]['lr']:.1e}",
            "time_s"     : round(elapsed, 1),
        })
        msg = (f"{stage_name} Ep {ep}/{epochs} | "
               f"train={train_loss:.4f} | val={val_loss:.4f} | {elapsed:.0f}s")
        print(msg, flush=True)
        log.info(msg)

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            torch.save({
                "epoch"            : ep,
                "stage"            : stage_name,
                "model_state_dict" : model.state_dict(),
                "val_loss"         : best_val_loss,
            }, save_path)
            log.info("  ✓ New best val_loss=%.4f saved.", best_val_loss)

    log.info("%s log:\n%s",
             stage_name, pd.DataFrame(rows).to_string(index=False))
    return best_val_loss


# ════════════════════════════════════════════════════════════
#  EVALUATION
#  Same metric set as YOLO and Mask R-CNN scripts.
# ════════════════════════════════════════════════════════════
def evaluate(model, loader):
    model.eval()

    map_metric    = MeanAveragePrecision(iou_type="bbox", class_metrics=True)

    # Segmentation accumulators (pixel-level, same formula as YOLO script)
    intersections = np.zeros(NUM_CLASSES)
    unions        = np.zeros(NUM_CLASSES)
    pred_areas    = np.zeros(NUM_CLASSES)
    target_areas  = np.zeros(NUM_CLASSES)

    # Per-class ROC / P-R-F1 accumulators
    y_true  = {c: [] for c in range(1, NUM_CLASSES)}
    y_score = {c: [] for c in range(1, NUM_CLASSES)}

    print("\n>>> Running evaluation pass...", flush=True)
    with torch.no_grad():
        for batch in tqdm(loader, desc="Evaluating", leave=True, dynamic_ncols=True):
            if not batch:
                continue
            imgs, gt_masks = batch        # gt_masks: (B, H, W) int64
            imgs = imgs.to(DEVICE, non_blocking=True)

            with torch.amp.autocast("cuda"):
                logits = model(imgs)      # (B, C, H, W)

            pred_maps = logits.argmax(dim=1).cpu().numpy()   # (B, H, W)
            gt_maps   = gt_masks.numpy()                     # (B, H, W)

            for pred_map, gt_map in zip(pred_maps, gt_maps):
                H, W = pred_map.shape

                # ── Segmentation pixel metrics ────────────────
                for c in range(1, NUM_CLASSES):
                    p_bin = (pred_map == c)
                    g_bin = (gt_map   == c)
                    intersections[c] += np.logical_and(p_bin, g_bin).sum()
                    unions[c]        += np.logical_or( p_bin, g_bin).sum()
                    pred_areas[c]    += p_bin.sum()
                    target_areas[c]  += g_bin.sum()

                # ── Post-process: semantic → instances ────────
                p_boxes, p_labels, p_scores, _ = postprocess_to_instances(pred_map)
                g_boxes, g_labels, g_scores, _ = postprocess_to_instances(gt_map, min_area=1)

                # Torchmetrics mAP update
                p_t = {
                    "boxes" : torch.tensor(p_boxes,  dtype=torch.float32)
                              if p_boxes else torch.zeros((0, 4)),
                    "scores": torch.tensor(p_scores, dtype=torch.float32)
                              if p_scores else torch.zeros(0),
                    "labels": torch.tensor(p_labels, dtype=torch.int64)
                              if p_labels else torch.zeros(0, dtype=torch.int64),
                }
                g_t = {
                    "boxes" : torch.tensor(g_boxes,  dtype=torch.float32)
                              if g_boxes else torch.zeros((0, 4)),
                    "labels": torch.tensor(g_labels, dtype=torch.int64)
                              if g_labels else torch.zeros(0, dtype=torch.int64),
                }
                map_metric.update([p_t], [g_t])

                # ── ROC / P-R-F1 accumulators ─────────────────
                p_boxes_t  = torch.tensor(p_boxes,  dtype=torch.float32) \
                             if p_boxes  else torch.zeros((0, 4))
                p_scores_t = torch.tensor(p_scores, dtype=torch.float32) \
                             if p_scores else torch.zeros(0)
                p_labels_t = torch.tensor(p_labels, dtype=torch.int64)   \
                             if p_labels else torch.zeros(0, dtype=torch.int64)
                g_boxes_t  = torch.tensor(g_boxes,  dtype=torch.float32) \
                             if g_boxes  else torch.zeros((0, 4))
                g_labels_t = torch.tensor(g_labels, dtype=torch.int64)   \
                             if g_labels else torch.zeros(0, dtype=torch.int64)

                for c in range(1, NUM_CLASSES):
                    cp_boxes  = p_boxes_t[ p_labels_t == c]
                    cp_scores = p_scores_t[p_labels_t == c]
                    ct_boxes  = g_boxes_t[ g_labels_t == c]

                    if len(ct_boxes) == 0:
                        for s in cp_scores:
                            y_score[c].append(s.item()); y_true[c].append(0)
                        continue
                    if len(cp_boxes) == 0:
                        for _ in ct_boxes:
                            y_score[c].append(0.0); y_true[c].append(1)
                        continue

                    ious        = box_iou(cp_boxes, ct_boxes)
                    matched_tgt = set()
                    for i, sc in enumerate(cp_scores):
                        max_iou, best_j = ious[i].max(0)
                        max_iou = max_iou.item(); best_j = best_j.item()
                        if max_iou >= IOU_THRESH and best_j not in matched_tgt:
                            y_score[c].append(sc.item()); y_true[c].append(1)
                            matched_tgt.add(best_j)
                        else:
                            y_score[c].append(sc.item()); y_true[c].append(0)
                    for _ in range(len(ct_boxes) - len(matched_tgt)):
                        y_score[c].append(0.0); y_true[c].append(1)

    # ── mAP ──────────────────────────────────────────────────
    map_res       = map_metric.compute()
    overall_map   = map_res["map"].item()
    overall_map50 = map_res["map_50"].item()
    overall_map75 = map_res["map_75"].item()
    per_cls_ap50  = map_res.get("map_per_class",
                                torch.zeros(NUM_CLASSES - 1))

    # ── Segmentation DataFrame ────────────────────────────────
    seg_rows = []
    for c in range(1, NUM_CLASSES):
        iou  = intersections[c] / (unions[c]                          + 1e-6)
        dice = (2 * intersections[c]) / (pred_areas[c] + target_areas[c] + 1e-6)
        seg_rows.append({
            "class": CLASS_NAMES[c],
            "mIoU" : round(float(iou),  4),
            "Dice" : round(float(dice), 4),
        })
    seg_df   = pd.DataFrame(seg_rows)
    macro_s  = pd.DataFrame([{
        "class": "Macro Average",
        "mIoU" : round(seg_df["mIoU"].mean(), 4),
        "Dice" : round(seg_df["Dice"].mean(), 4),
    }])
    seg_df   = pd.concat([seg_df, macro_s], ignore_index=True)

    # ── Detection P/R/F1 + AUC DataFrame ─────────────────────
    det_rows = []
    roc_data = {}
    for c in range(1, NUM_CLASSES):
        yt = np.array(y_true[c]); ys = np.array(y_score[c])
        if len(yt) == 0 or yt.sum() == 0:
            det_rows.append({"class": CLASS_NAMES[c], "AP@0.50": 0.0,
                             "Precision": 0.0, "Recall": 0.0,
                             "F1": 0.0, "AUC": float("nan")})
            continue

        fpr, tpr, _ = roc_curve(yt, ys)
        roc_auc     = auc(fpr, tpr)
        roc_data[c] = (fpr, tpr, roc_auc)

        yp   = (ys >= SCORE_THRESH).astype(int)
        tp   = int(((yp == 1) & (yt == 1)).sum())
        fp   = int(((yp == 1) & (yt == 0)).sum())
        fn   = int(((yp == 0) & (yt == 1)).sum())
        prec = tp / (tp + fp + 1e-9)
        rec  = tp / (tp + fn + 1e-9)
        f1   = 2 * prec * rec / (prec + rec + 1e-9)
        ap50 = float(per_cls_ap50[c - 1]) if c - 1 < len(per_cls_ap50) else 0.0

        det_rows.append({
            "class"    : CLASS_NAMES[c],
            "AP@0.50"  : round(ap50,    4),
            "Precision": round(prec,    4),
            "Recall"   : round(rec,     4),
            "F1"       : round(f1,      4),
            "AUC"      : round(roc_auc, 4),
        })

    det_df  = pd.DataFrame(det_rows)
    macro_d = pd.DataFrame([{
        "class"    : "Macro Average",
        "AP@0.50"  : round(det_df["AP@0.50"].mean(),   4),
        "Precision": round(det_df["Precision"].mean(), 4),
        "Recall"   : round(det_df["Recall"].mean(),    4),
        "F1"       : round(det_df["F1"].mean(),        4),
        "AUC"      : round(det_df["AUC"].mean(),       4),
    }])
    det_df  = pd.concat([det_df, macro_d], ignore_index=True)

    # ── Overview DataFrame ────────────────────────────────────
    overview_df = pd.DataFrame([
        {"metric": "mAP @ [0.50:0.95]", "detection": round(overall_map,  4), "segmentation": ""},
        {"metric": "mAP @ 0.50",         "detection": round(overall_map50,4), "segmentation": ""},
        {"metric": "mAP @ 0.75",         "detection": round(overall_map75,4), "segmentation": ""},
        {"metric": "Macro Precision",     "detection": round(det_df.iloc[:-1]["Precision"].mean(), 4), "segmentation": ""},
        {"metric": "Macro Recall",        "detection": round(det_df.iloc[:-1]["Recall"].mean(),    4), "segmentation": ""},
        {"metric": "Macro F1",            "detection": round(det_df.iloc[:-1]["F1"].mean(),        4), "segmentation": ""},
        {"metric": "Macro AUC",           "detection": round(det_df.iloc[:-1]["AUC"].mean(),       4), "segmentation": ""},
        {"metric": "Macro mIoU",          "detection": "",  "segmentation": round(seg_df.iloc[:-1]["mIoU"].mean(), 4)},
        {"metric": "Macro Dice",          "detection": "",  "segmentation": round(seg_df.iloc[:-1]["Dice"].mean(), 4)},
    ])

    # ── ROC plot ──────────────────────────────────────────────
    n  = NUM_CLASSES - 1
    fig, axes = plt.subplots(1, n, figsize=(5 * n, 4))
    for ci, ax in enumerate(axes, start=1):
        if ci in roc_data:
            fpr, tpr, roc_auc = roc_data[ci]
            ax.plot(fpr, tpr, lw=2, label=f"AUC={roc_auc:.3f}")
            ax.plot([0, 1], [0, 1], "k--", lw=1)
            ax.legend()
        ax.set_title(CLASS_NAMES[ci])
        ax.set_xlabel("FPR"); ax.set_ylabel("TPR")
        ax.set_xlim(0, 1);    ax.set_ylim(0, 1)
    plt.suptitle("ROC Curves — U-Net Scratch Detection (per class)", fontsize=14)
    plt.tight_layout()
    plt.savefig(ROC_PATH, dpi=120, bbox_inches="tight")
    plt.show()
    log.info("ROC figure saved → %s", ROC_PATH)

    return overview_df, det_df, seg_df


# ════════════════════════════════════════════════════════════
#  MAIN
# ════════════════════════════════════════════════════════════
def main():
    model = build_model(pretrained=False)  # random weight initialisation — true scratch training
    total = sum(p.numel() for p in model.parameters())
    log.info("U-Net ResNet-34 encoder  |  total params: %s", f"{total:,}")
    log.info("Strategy: FROM SCRATCH — all weights randomly initialised")

    criterion     = SegLoss(NUM_CLASSES, ce_weight=0.5).to(DEVICE)
    best_val_loss = float("inf")

    # ── Single stage: train all layers end-to-end from random init ──
    # No frozen warmup — there are no pretrained features to protect.
    # All layers receive gradients from epoch 1.
    log.info("No frozen stage — training all layers from random init.")
    print("Training all layers from scratch — no frozen stage.", flush=True)
    best_val_loss = train_stage(
        model, train_loader, val_loader, criterion,
        epochs=STAGE2_EPOCHS, lr=STAGE2_LR,
        stage_name="Scratch-FullTraining",
        best_val_loss=best_val_loss, save_path=SAVE_PATH,
    )

    log.info("Training complete. Loading best checkpoint...")
    ckpt = torch.load(SAVE_PATH, map_location=DEVICE)
    model.load_state_dict(ckpt["model_state_dict"])
    log.info("Best checkpoint: stage=%s  epoch=%d  val_loss=%.4f",
             ckpt["stage"], ckpt["epoch"], ckpt["val_loss"])

    # ── Evaluation ───────────────────────────────────────────
    overview_df, det_df, seg_df = evaluate(model, val_loader)

    print("\n" + "="*60, flush=True)
    print("  FINAL METRICS SUMMARY", flush=True)
    print("="*60, flush=True)
    print("\nOverview:", flush=True)
    print(overview_df.to_string(index=False), flush=True)
    print("\nDetection:", flush=True)
    print(det_df.to_string(index=False), flush=True)
    print("\nSegmentation:", flush=True)
    print(seg_df.to_string(index=False), flush=True)
    log.info("=== FINAL METRICS SUMMARY ===")
    log.info("Overview:\n%s",     overview_df.to_string(index=False))
    log.info("Detection:\n%s",    det_df.to_string(index=False))
    log.info("Segmentation:\n%s", seg_df.to_string(index=False))
    log.info("Best model → %s",   SAVE_PATH)
    log.info("ROC curves → %s",   ROC_PATH)


if __name__ == "__main__":
    main()
```
