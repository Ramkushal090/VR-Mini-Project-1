```python
# ============================================================
#  YOLO Transfer Learning — DeepFashion2 Top-5 Categories
#  Master-JSON Dataset  |  Dual-GPU (T4×2)
#  Task 3.2: Detection + Segmentation
# ============================================================

# ─── 0. INSTALLATION ────────────────────────────────────────
import subprocess, sys

def _install(pkg: str) -> None:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", pkg])

for _pkg in [
    "ultralytics", "scikit-learn", "matplotlib", "seaborn",
    "Pillow", "tqdm", "pyyaml", "opencv-python-headless", "pandas",
]:
    _install(_pkg)

# ─── 1. STDLIB / THIRD-PARTY IMPORTS ────────────────────────
import json
import logging
import multiprocessing as mp
import random
import shutil
import time
import warnings
from collections import defaultdict
from concurrent.futures import ProcessPoolExecutor, as_completed
from pathlib import Path

import cv2
import matplotlib.patches as patches
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import torch
import yaml
from PIL import Image
from sklearn.metrics import auc, roc_curve
from tqdm import tqdm
from ultralytics import YOLO

warnings.filterwarnings("ignore", category=FutureWarning)

# ─── 2. LOGGING SETUP ───────────────────────────────────────
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s  [%(levelname)-8s]  %(name)s — %(message)s",
    datefmt="%H:%M:%S",
)
ROOT_LOG = logging.getLogger("yolo_fashion")


# ════════════════════════════════════════════════════════════
#  CLASS 1 — DatasetConfig
#  Central configuration hub: paths, class mappings,
#  hyperparameters, and dataset.yaml generation.
# ════════════════════════════════════════════════════════════
class DatasetConfig:
    """
    Stores every configurable constant in one place.

    Class mapping contract
    ─────────────────────
    Master JSON ``top_5_categories`` field: [1, 8, 7, 2, 9]
    These DeepFashion2 category IDs are assigned YOLO 0-based
    indices in the order they appear in that list:

        DF2 cat_id   →  YOLO idx  →  class name
        ──────────────────────────────────────────
             1       →     0      →  short_sleeve_top
             8       →     1      →  trousers
             7       →     2      →  shorts
             2       →     3      →  long_sleeve_top
             9       →     4      →  skirt

    Semantic mask convention (identical to original script):
        pixel 0  = background (unannotated)
        pixel N  = foreground class N  (= yolo_idx + 1)
    """

    # Full DF2 registry — used for human-readable names only
    _DF2_REGISTRY: dict = {
        1:  "short_sleeve_top",
        2:  "long_sleeve_top",
        3:  "short_sleeve_outwear",
        4:  "long_sleeve_outwear",
        5:  "vest",
        6:  "sling",
        7:  "shorts",
        8:  "trousers",
        9:  "skirt",
        10: "short_sleeve_dress",
        11: "long_sleeve_dress",
        12: "vest_dress",
        13: "sling_dress",
    }

    def __init__(self) -> None:
        self._log = logging.getLogger(f"{ROOT_LOG.name}.config")

        # ── Paths ───────────────────────────────────────────
        _base = Path(
            "/kaggle/input/datasets/hrushi1998"
            "/vr-mini-proj-dataset/Trimmed_Dataset"
        )

        # Training JSON is already subsampled externally — every record
        # in this file is converted directly; no internal random sampling.
        self.train_json_path: Path = Path(
            "/kaggle/input/datasets/hades1998/subsampling-json/subsampled_train_data.json"
        )

        # Validation uses the full trimmed val set from the main dataset.
        self.val_json_path: Path = (
            _base / "trimmed_val_data" / "processed_val_data.json"
        )

        # Image base directories — the JSON and the images are in different
        # locations for train (subsampled JSON is a separate dataset), so we
        # track them explicitly instead of deriving from json_path.parent.
        self.img_base: dict = {
            "train"      : _base / "trimmed_train_data",
            "validation" : _base / "trimmed_val_data",
        }
        self.work_dir:    Path = Path("/kaggle/working/yolo_fashion_oop")
        self.dataset_dir: Path = self.work_dir / "dataset"
        self.results_dir: Path = self.work_dir / "results"

        for _d in [self.work_dir, self.dataset_dir, self.results_dir]:
            _d.mkdir(parents=True, exist_ok=True)

        # ── Class Mapping ────────────────────────────────────
        # Order is intentional — matches the master JSON field.
        self.top5_cat_ids: list = [1, 8, 7, 2, 9]

        # DF2 category_id  →  YOLO 0-based index
        self.cat_id_to_yolo: dict = {
            cat_id: yolo_idx
            for yolo_idx, cat_id in enumerate(self.top5_cat_ids)
        }

        # YOLO 0-based index  →  human-readable name
        self.yolo_to_name: dict = {
            yolo_idx: self._DF2_REGISTRY[cat_id]
            for yolo_idx, cat_id in enumerate(self.top5_cat_ids)
        }

        self.num_classes: int = len(self.top5_cat_ids)

        # ── GPU Setup ────────────────────────────────────────
        self.num_gpus    = torch.cuda.device_count()
        self.gpu_names   = [torch.cuda.get_device_name(i) for i in range(self.num_gpus)]
        self.device_str  = (
            ",".join(str(i) for i in range(self.num_gpus))
            if self.num_gpus > 0 else "cpu"
        )
        self.infer_device  = "0" if self.num_gpus > 0 else "cpu"
        self.num_workers   = min(8, mp.cpu_count())
        self.prep_workers  = max(1, mp.cpu_count() - 1)

        # ── Model ────────────────────────────────────────────
        self.base_weights   = "yolov8n-seg.pt"
        self.infer_batch_sz = 16

        
        self.train_hp: dict = {
            "epochs"        : 20,
            "imgsz"         : 640,
            "batch"         : 16,       # 8 per GPU on T4×2
            "workers"       : self.num_workers,
            "device"        : self.device_str,
            "patience"      : 5,
            "lr0"           : 0.005,
            "lrf"           : 0.01,
            "momentum"      : 0.937,
            "weight_decay"  : 0.0005,
            "warmup_epochs" : 1,
            "cos_lr"        : True,
            "amp"           : True,     # FP16 — ~2× speedup, halves VRAM
            "cache"         : "disk",   # decoded images → /tmp, not RAM
            "augment"       : True,
            "mosaic"        : 1.0,
            "mixup"         : 0.1,
            "copy_paste"    : 0.0,      # disabled — high memory cost
            "degrees"       : 10.0,
            "translate"     : 0.1,
            "scale"         : 0.5,
            "fliplr"        : 0.5,
            "hsv_h"         : 0.015,
            "hsv_s"         : 0.7,
            "hsv_v"         : 0.4,
            "verbose"       : True,
            "plots"         : True,
            "save"          : True,
            "project"       : str(self.results_dir),
            "name"          : "yolo_fashion_oop",
            "exist_ok"      : True,
            "pretrained"    : True,     # ← Transfer Learning from COCO
            "freeze"        : 10,       # freeze first 10 backbone layers
        }

        self.yaml_path: Path = self.dataset_dir / "dataset.yaml"

    # ── Public helpers ────────────────────────────────────────

    def generate_yaml(self) -> Path:
        """Write dataset.yaml required by YOLO trainer."""
        cfg = {
            "path"  : str(self.dataset_dir),
            "train" : "images/train",
            "val"   : "images/validation",
            "nc"    : self.num_classes,
            "names" : [self.yolo_to_name[i] for i in range(self.num_classes)],
        }
        with open(self.yaml_path, "w") as fh:
            yaml.dump(cfg, fh, default_flow_style=False)
        self._log.info("dataset.yaml → %s", self.yaml_path)
        self._log.info("Classes (%d): %s", self.num_classes, cfg["names"])
        return self.yaml_path

    def log_summary(self) -> None:
        """Emit a tidy DataFrame of the active configuration."""
        self._log.info(
            "GPU inventory: %d device(s)  %s",
            self.num_gpus, self.gpu_names,
        )
        self._log.info(
            "Devices — train: %s  |  infer: %s",
            self.device_str, self.infer_device,
        )
        self._log.info(
            "CPU cores: %d  (prep_workers=%d, data_workers=%d)",
            mp.cpu_count(), self.prep_workers, self.num_workers,
        )

        class_df = pd.DataFrame(
            [
                {
                    "yolo_idx" : yi,
                    "df2_cat"  : self.top5_cat_ids[yi],
                    "sem_pixel": yi + 1,
                    "name"     : self.yolo_to_name[yi],
                }
                for yi in range(self.num_classes)
            ]
        )
        self._log.info("Class mapping:\n%s", class_df.to_string(index=False))

        hp_df = pd.DataFrame(
            list(self.train_hp.items()), columns=["parameter", "value"]
        )
        self._log.info("Training hyperparameters:\n%s", hp_df.to_string(index=False))


# ════════════════════════════════════════════════════════════
#  MODULE-LEVEL WORKER  (must be picklable — defined outside
#  any class so ProcessPoolExecutor can serialise it cleanly)
# ════════════════════════════════════════════════════════════
def _convert_record(args: tuple) -> bool:
    """
    ProcessPoolExecutor worker: convert one master-JSON record to YOLO format.

    Parameters (packed into a single tuple for imap compatibility)
    ─────────────────────────────────────────────────────────────
    rec            : dict  — single entry from master JSON ``data`` list
    img_base_str   : str   — directory that contains the images
    cat_id_to_yolo : dict  — DF2 category_id → YOLO 0-based index
    img_out_str    : str   — destination for image copy
    lbl_out_str    : str   — destination for YOLO .txt label
    mask_out_str   : str   — destination for semantic .png mask

    Returns True on success, False when the record is skipped.

    Outputs per record
    ──────────────────
    • <stem>.txt   — YOLO polygon label (class_id + normalised coords)
    • <stem>.png   — semantic mask
                     pixel 0 = background
                     pixel N = foreground (N = yolo_idx + 1)
    • <stem>.jpg   — image copy (hard-linked when possible)
    """
    rec, img_base_str, cat_id_to_yolo, img_out_str, lbl_out_str, mask_out_str = args

    img_base  = Path(img_base_str)
    img_out   = Path(img_out_str)
    lbl_out   = Path(lbl_out_str)
    mask_out  = Path(mask_out_str)

    # ── Locate source image ──────────────────────────────────
    rel_path = rec.get("image_path", "")
    img_fp   = img_base / rel_path
    if not img_fp.exists():
        return False

    # ── Filter items to top-5 classes only ──────────────────
    item_cats  = rec.get("item_categories",      [])
    seg_polys  = rec.get("segmentation_polygons", [])

    valid_items = []
    for idx, cat_id in enumerate(item_cats):
        if cat_id not in cat_id_to_yolo:
            continue
        # segmentation_polygons[idx] is a list of polygon rings;
        # each ring is a flat list [x, y, x, y, …]
        rings    = seg_polys[idx] if idx < len(seg_polys) else [[]]
        raw_poly = rings[0] if rings else []
        valid_items.append(
            {"yolo_cls": cat_id_to_yolo[cat_id], "polygon": raw_poly}
        )

    if not valid_items:
        return False

    # ── Open image to get dimensions ─────────────────────────
    try:
        pil_img      = Image.open(img_fp)
        img_w, img_h = pil_img.size
    except Exception:
        return False

    sem_mask    = np.zeros((img_h, img_w), dtype=np.uint8)
    label_lines = []

    for item in valid_items:
        yolo_cls = item["yolo_cls"]
        sem_idx  = yolo_cls + 1          # 1-based semantic pixel value
        raw_poly = item["polygon"]       # flat [x, y, x, y, …]

        if len(raw_poly) < 6:            # need at least 3 coordinate pairs
            continue

        # Rasterise polygon into semantic mask (CPU / OpenCV)
        pts_abs = [
            [int(round(raw_poly[i])), int(round(raw_poly[i + 1]))]
            for i in range(0, len(raw_poly) - 1, 2)
        ]
        pts_np = np.array(pts_abs, dtype=np.int32).reshape((-1, 1, 2))
        pts_np[..., 0] = np.clip(pts_np[..., 0], 0, img_w - 1)
        pts_np[..., 1] = np.clip(pts_np[..., 1], 0, img_h - 1)
        cv2.fillPoly(sem_mask, [pts_np], int(sem_idx))

        # Normalise polygon to [0, 1] for YOLO label
        norm_coords: list = []
        for i in range(0, len(raw_poly) - 1, 2):
            norm_coords.append(max(0.0, min(1.0, raw_poly[i]     / img_w)))
            norm_coords.append(max(0.0, min(1.0, raw_poly[i + 1] / img_h)))

        pts_str = " ".join(f"{v:.6f}" for v in norm_coords)
        label_lines.append(f"{yolo_cls} {pts_str}")

    if not label_lines:
        return False

    stem = img_fp.stem

    # ── Write outputs ────────────────────────────────────────
    (lbl_out / f"{stem}.txt").write_text("\n".join(label_lines))
    Image.fromarray(sem_mask).save(mask_out / f"{stem}.png")

    dst = img_out / img_fp.name
    if not dst.exists():
        shutil.copy2(img_fp, dst)

    return True


# ════════════════════════════════════════════════════════════
#  CLASS 2 — YoloDataBuilder
#  Parses Master JSONs, converts records to YOLO format, and
#  generates binary semantic PNG masks.
#  Parallelism: concurrent.futures.ProcessPoolExecutor
# ════════════════════════════════════════════════════════════
class YoloDataBuilder:
    """
    Reads the master JSON for each split and writes:

        dataset/
          images/<split>/      — copied source images
          labels/<split>/      — YOLO polygon .txt  (0-indexed, normalised)
          seg_masks/<split>/   — semantic .png masks
                                 pixel 0 = background
                                 pixel N = class N  (N = yolo_idx + 1)

    Parallel workers are spawned via ``ProcessPoolExecutor`` so that
    JSON parsing, polygon rasterisation, and file I/O saturate all
    available CPU cores simultaneously.  The structural footprint
    differs from the original ``multiprocessing.Pool`` approach:
    futures are collected with ``as_completed``, allowing the progress
    bar to update as each task finishes rather than in submission order.
    """

    def __init__(self, config: DatasetConfig) -> None:
        self.cfg  = config
        self._log = logging.getLogger(f"{ROOT_LOG.name}.builder")

    def build_all(self) -> tuple:
        """Build both splits; return (n_train, n_val)."""
        n_train = self._build_split(self.cfg.train_json_path, "train")
        n_val   = self._build_split(self.cfg.val_json_path,   "validation")
        self._log.info(
            "Dataset build complete.  train=%d  val=%d", n_train, n_val
        )
        return n_train, n_val

    def compute_class_weights(self) -> np.ndarray:
        """
        Inverse-frequency class weights from training label files.
        Identical formula to the original script; output via DataFrame.
        """
        lbl_dir = self.cfg.dataset_dir / "labels" / "train"
        counts  = np.zeros(self.cfg.num_classes, dtype=np.float64)

        for fp in lbl_dir.glob("*.txt"):
            for line in fp.read_text().splitlines():
                parts = line.strip().split()
                if parts:
                    counts[int(parts[0])] += 1

        counts  = np.where(counts == 0, 1.0, counts)
        weights = 1.0 / counts
        weights = weights / weights.sum() * self.cfg.num_classes

        df = pd.DataFrame(
            {
                "yolo_idx"  : range(self.cfg.num_classes),
                "class_name": [self.cfg.yolo_to_name[i] for i in range(self.cfg.num_classes)],
                "count"     : counts,
                "weight"    : weights,
            }
        )
        self._log.info(
            "Class weights (inverse-frequency):\n%s", df.round(4).to_string(index=False)
        )
        return weights

    # ── Private helpers ───────────────────────────────────────

    def _build_split(self, json_path: Path, split_name: str) -> int:
        """Parse one master JSON and write YOLO files for that split."""
        img_out  = self.cfg.dataset_dir / "images"    / split_name
        lbl_out  = self.cfg.dataset_dir / "labels"    / split_name
        mask_out = self.cfg.dataset_dir / "seg_masks" / split_name
        for _d in [img_out, lbl_out, mask_out]:
            _d.mkdir(parents=True, exist_ok=True)

        with open(json_path) as fh:
            master_json = json.load(fh)

        records  = master_json["data"]
        # Use the explicit image base for this split — the subsampled train
        # JSON lives in a different Kaggle dataset from the actual images.
        img_base = self.cfg.img_base[split_name]

        self._log.info(
            "[%s] %d records loaded from %s", split_name, len(records), json_path
        )

        # Pack args for the module-level worker
        worker_args = [
            (
                rec,
                str(img_base),
                self.cfg.cat_id_to_yolo,
                str(img_out),
                str(lbl_out),
                str(mask_out),
            )
            for rec in records
        ]

        written = skipped = 0

        # ProcessPoolExecutor — as_completed lets the progress bar
        # update as individual futures finish (non-submission order).
        with ProcessPoolExecutor(max_workers=self.cfg.prep_workers) as executor:
            future_map = {
                executor.submit(_convert_record, arg): arg
                for arg in worker_args
            }
            with tqdm(
                total=len(future_map),
                desc=f"  Building [{split_name}]  ({self.cfg.prep_workers} workers)",
            ) as pbar:
                for future in as_completed(future_map):
                    if future.result():
                        written += 1
                    else:
                        skipped += 1
                    pbar.update(1)

        self._log.info(
            "[%s] written=%d  skipped=%d", split_name, written, skipped
        )
        return written


# ════════════════════════════════════════════════════════════
#  CLASS 3 — YoloTrainer
#  Dual-GPU training execution and weight management.
# ════════════════════════════════════════════════════════════
class YoloTrainer:
    """
    Wraps ``YOLO.train()`` for dual-GPU (DataParallel) execution.

    Transfer-learning strategy
    ──────────────────────────
    • Weights   : ``yolov8n-seg.pt``  (COCO pretrained)
    • Frozen    : first ``freeze=10`` backbone layers
    • Device    : ``"0,1"`` → torch.nn.DataParallel across both T4s
    • AMP       : FP16 mixed precision (``amp=True``)
    • Cache     : decoded images written to /tmp (``cache="disk"``)
    """

    def __init__(self, config: DatasetConfig) -> None:
        self.cfg      = config
        self._log     = logging.getLogger(f"{ROOT_LOG.name}.trainer")
        self.model    = None
        self.save_dir: Path = None

    def load_pretrained(self) -> YOLO:
        """
        Instantiate YOLOv8n-seg from COCO pretrained weights and
        register epoch-timing callbacks so wall-clock time per epoch
        is captured on the main process even under DataParallel.
        """
        self.model = YOLO(self.cfg.base_weights)

        # Accumulators written by the callbacks below
        self._epoch_times: list  = []   # duration (seconds) per completed epoch
        self._epoch_start: float = 0.0  # timestamp set at epoch start

        # YOLO callbacks fire on the main process for both single-GPU
        # and multi-GPU DataParallel runs.
        self.model.add_callback("on_train_epoch_start", self._cb_epoch_start)
        self.model.add_callback("on_train_epoch_end",   self._cb_epoch_end)

        n_params = sum(p.numel() for p in self.model.model.parameters())
        self._log.info(
            "Loaded %s  (%s parameters)",
            self.cfg.base_weights, f"{n_params:,}",
        )
        self._log.info(
            "Strategy: COCO → top-5 transfer, %d backbone layers frozen",
            self.cfg.train_hp["freeze"],
        )
        self._log.info("Epoch-timing callbacks registered.")
        return self.model

    def train(self, yaml_path: Path) -> Path:
        """
        Execute training; return path to the saved best weights.

        The training configuration is emitted as a pandas DataFrame
        before training starts so TAs can inspect every hyperparameter
        without wading through log noise.
        """
        hp = self.cfg.train_hp
        hp_df = pd.DataFrame(list(hp.items()), columns=["parameter", "value"])
        self._log.info("Training configuration:\n%s", hp_df.to_string(index=False))

        t_start = time.time()

        results = self.model.train(data=str(yaml_path), **hp)

        elapsed_s = time.time() - t_start
        self._log.info(
            "Training finished in %.1fs  (%dm %ds)",
            elapsed_s, int(elapsed_s // 60), int(elapsed_s % 60),
        )

        self.save_dir = self._resolve_save_dir(results)
        best_pt       = self.save_dir / "weights" / "best.pt"

        if not best_pt.exists():
            last_pt = self.save_dir / "weights" / "last.pt"
            if last_pt.exists():
                best_pt = last_pt
                self._log.warning("best.pt absent — falling back to last.pt")

        self._log.info("Save dir   : %s", self.save_dir)
        self._log.info("Best model : %s", best_pt)
        self._log_epoch_table()
        return best_pt

    # ── Private helpers ───────────────────────────────────────

    def _resolve_save_dir(self, results) -> Path:
        """
        Locate the YOLO output directory robustly.
        Under multi-GPU DataParallel the ``results`` object returned by
        ``model.train()`` is often None on the main process; we fall
        through three resolution strategies before raising.
        """
        if results is not None and hasattr(results, "save_dir"):
            return Path(results.save_dir)

        if getattr(self.model, "trainer", None) is not None:
            if hasattr(self.model.trainer, "save_dir"):
                return Path(self.model.trainer.save_dir)

        named = Path(self.cfg.train_hp["project"]) / self.cfg.train_hp["name"]
        if named.exists():
            return named

        candidates = [
            d for d in Path(self.cfg.train_hp["project"]).iterdir()
            if d.is_dir()
        ]
        if candidates:
            return max(candidates, key=lambda d: d.stat().st_mtime)

        raise FileNotFoundError(
            f"Cannot locate training output under {self.cfg.train_hp['project']}"
        )

    # ── Epoch-timing callbacks ────────────────────────────────

    def _cb_epoch_start(self, trainer) -> None:
        """Record wall-clock time at the beginning of each epoch."""
        self._epoch_start = time.time()

    def _cb_epoch_end(self, trainer) -> None:
        """
        Compute elapsed time for the finished epoch, store it, and
        log an immediate one-liner so the duration appears in the
        notebook output right after YOLO's own epoch summary line.
        """
        duration = time.time() - self._epoch_start
        self._epoch_times.append(duration)
        self._log.info(
            "Epoch %d/%d  —  wall-clock time: %.1fs  (%dm %ds)",
            trainer.epoch + 1,
            trainer.epochs,
            duration,
            int(duration // 60),
            int(duration  % 60),
        )

    def _log_epoch_table(self) -> None:
        """
        Parse results.csv, stitch in the per-epoch wall-clock times
        collected by the callbacks, and emit the full table as a
        pandas DataFrame so every epoch's losses, mAP, and duration
        are visible in one place.
        """
        csv_path = self.save_dir / "results.csv"
        if not csv_path.exists():
            self._log.warning("results.csv not found — epoch log unavailable")
            return

        df = pd.read_csv(csv_path)
        df.columns = [c.strip() for c in df.columns]

        # Attach per-epoch durations when the callbacks collected them.
        # min() guards against the rare case where early-stopping caused
        # fewer epochs to complete than rows written to results.csv.
        if self._epoch_times:
            n_rows = min(len(df), len(self._epoch_times))
            df.loc[: n_rows - 1, "epoch_time_s"] = [
                round(t, 1) for t in self._epoch_times[:n_rows]
            ]
            total_s = sum(self._epoch_times)
            self._log.info(
                "Total epoch wall-clock time: %.1fs  (%dm %ds)  "
                "| mean per epoch: %.1fs",
                total_s,
                int(total_s // 60),
                int(total_s  % 60),
                total_s / len(self._epoch_times),
            )

        self._log.info(
            "Epoch-by-epoch training log (epoch_time_s = wall-clock seconds):\n%s",
            df.round(4).to_string(index=False),
        )


# ════════════════════════════════════════════════════════════
#  CLASS 4 — YoloEvaluator
#  Batched GPU inference + detection / segmentation metrics.
# ════════════════════════════════════════════════════════════
class YoloEvaluator:
    """
    Runs batched GPU inference over the validation split and
    computes all required metrics.

    Metric inventory (mathematically identical to source script)
    ────────────────────────────────────────────────────────────
    Detection
      • mAP @ 0.50, 0.75, [0.50:0.95]  (11-point interpolated AP)
      • Precision, Recall, F1  (conf≥0.25, IoU≥0.50)
      • ROC curve + AUC  (per class, image-level binary classification)

    Segmentation  (foreground-only; background pixel=0 excluded)
      • mIoU  (mask intersection-over-union)
      • Dice coefficient  (2·|P∩G| / (|P|+|G|))

    All final summaries are printed as pandas DataFrames.
    """

    IOU_THRESHOLDS: np.ndarray = np.arange(0.5, 1.0, 0.05)

    def __init__(self, config: DatasetConfig, best_model_path: Path) -> None:
        self.cfg  = config
        self._log = logging.getLogger(f"{ROOT_LOG.name}.evaluator")

        self.model = YOLO(str(best_model_path))
        self._log.info("Best model loaded from %s", best_model_path)

        self.val_img_dir  = config.dataset_dir / "images"    / "validation"
        self.val_lbl_dir  = config.dataset_dir / "labels"    / "validation"
        self.val_mask_dir = config.dataset_dir / "seg_masks" / "validation"

        self.val_images: list = (
            sorted(self.val_img_dir.glob("*.jpg"))
            + sorted(self.val_img_dir.glob("*.png"))
        )
        self._log.info(
            "Validation images available: %d", len(self.val_images)
        )

        self._reset_accumulators()

        # DataFrames populated by compute_* methods
        self._det_df: pd.DataFrame = None
        self._seg_df: pd.DataFrame = None
        self._roc_df: pd.DataFrame = None

    # ── Public interface ──────────────────────────────────────

    def run_inference(self) -> None:
        """
        Batched GPU inference.
        INFER_BATCH_SIZE images are passed per ``model.predict()`` call
        so the GPU stays fully loaded between calls, mirroring the
        original script's batching strategy.
        """
        bs       = self.cfg.infer_batch_sz
        batches  = [
            self.val_images[i: i + bs]
            for i in range(0, len(self.val_images), bs)
        ]
        self._log.info(
            "Batched GPU inference — %d images, batch_size=%d, device=%s",
            len(self.val_images), bs, self.cfg.infer_device,
        )

        for batch_fps in tqdm(batches, desc="GPU inference"):
            batch_preds = self.model.predict(
                source  = [str(fp) for fp in batch_fps],
                imgsz   = self.cfg.train_hp["imgsz"],
                device  = self.cfg.infer_device,
                verbose = False,
                conf    = 0.1,       # kept low; filtered per metric below
                stream  = False,     # return all at once — faster for batch
            )
            for img_fp, res in zip(batch_fps, batch_preds):
                self._accumulate_image(img_fp, res)

        self._log.info("Inference and metric accumulation complete.")

    def compute_detection_metrics(self) -> pd.DataFrame:
        """
        Compute per-class AP table, then P/R/F1 at conf≥0.25, IoU≥0.50.
        Returns a pandas DataFrame (last row = macro average).
        """
        n        = self.cfg.num_classes
        ap_table = np.zeros((n, len(self.IOU_THRESHOLDS)))

        for ci in range(n):
            for ti, iou_t in enumerate(self.IOU_THRESHOLDS):
                ap_table[ci, ti] = self._compute_ap(
                    self.per_class_preds[ci],
                    self.per_class_gt_cnt[ci],
                    iou_t,
                )

        self._ap_table = ap_table          # cache for final summary
        ap75_idx = min(5, len(self.IOU_THRESHOLDS) - 1)

        rows = []
        for ci in range(n):
            P, R, F1 = self._prf_at_conf(
                self.per_class_preds[ci], self.per_class_gt_cnt[ci]
            )
            rows.append(
                {
                    "class"        : self.cfg.yolo_to_name[ci],
                    "AP@0.50"      : round(ap_table[ci, 0],       4),
                    "AP@0.75"      : round(ap_table[ci, ap75_idx], 4),
                    "mAP@[.5:.95]" : round(ap_table[ci].mean(),   4),
                    "Precision"    : round(P,  4),
                    "Recall"       : round(R,  4),
                    "F1"           : round(F1, 4),
                }
            )

        df = pd.DataFrame(rows)
        macro = {
            "class"        : "Macro Average",
            "AP@0.50"      : round(df["AP@0.50"].mean(),      4),
            "AP@0.75"      : round(df["AP@0.75"].mean(),      4),
            "mAP@[.5:.95]" : round(df["mAP@[.5:.95]"].mean(),4),
            "Precision"    : round(df["Precision"].mean(),    4),
            "Recall"       : round(df["Recall"].mean(),       4),
            "F1"           : round(df["F1"].mean(),           4),
        }
        self._det_df = pd.concat(
            [df, pd.DataFrame([macro])], ignore_index=True
        )
        self._log.info(
            "Detection metrics:\n%s", self._det_df.to_string(index=False)
        )
        return self._det_df

    def compute_segmentation_metrics(self) -> pd.DataFrame:
        """
        Per-class mIoU and Dice over foreground pixels only.
        Returns a pandas DataFrame (last row = macro average).
        """
        rows = []
        for ci in range(self.cfg.num_classes):
            ious  = self.sem_iou_per_class[ci]
            dices = self.sem_dice_per_class[ci]
            rows.append(
                {
                    "class"    : self.cfg.yolo_to_name[ci],
                    "sem_pixel": ci + 1,
                    "mIoU"     : round(np.mean(ious)  if ious  else 0.0, 4),
                    "Dice"     : round(np.mean(dices) if dices else 0.0, 4),
                    "n_images" : len(ious),
                }
            )

        df = pd.DataFrame(rows)
        macro = {
            "class"    : "Macro Average",
            "sem_pixel": "—",
            "mIoU"     : round(df["mIoU"].mean(), 4),
            "Dice"     : round(df["Dice"].mean(), 4),
            "n_images" : "—",
        }
        self._seg_df = pd.concat(
            [df, pd.DataFrame([macro])], ignore_index=True
        )
        self._log.info(
            "Segmentation metrics (background pixel=0 excluded):\n%s",
            self._seg_df.to_string(index=False),
        )
        return self._seg_df

    def compute_roc_auc(self, output_dir: Path) -> pd.DataFrame:
        """
        Compute ROC curves and AUC for each class (image-level binary).
        Saves the figure to ``output_dir/roc_curves_detection.png``.
        Returns a pandas DataFrame including macro-average AUC.
        """
        n   = self.cfg.num_classes
        fig, axes = plt.subplots(1, n, figsize=(5 * n, 4))
        axes = [axes] if n == 1 else list(axes)

        rows = []
        for ci in range(n):
            yt   = np.array(self.y_true_det[ci])
            ys   = np.array(self.y_score_det[ci])
            ax   = axes[ci]
            name = self.cfg.yolo_to_name[ci]

            if yt.sum() == 0 or (1 - yt).sum() == 0:
                roc_auc = float("nan")
                ax.set_title(f"{name}\n(AUC undefined)")
            else:
                fpr, tpr, _ = roc_curve(yt, ys)
                roc_auc     = auc(fpr, tpr)
                ax.plot(fpr, tpr, lw=2, label=f"AUC = {roc_auc:.3f}")
                ax.plot([0, 1], [0, 1], "k--", lw=1)
                ax.set_xlabel("FPR")
                ax.set_ylabel("TPR")
                ax.legend()
                ax.set_xlim(0, 1)
                ax.set_ylim(0, 1)
                ax.set_title(name)

            rows.append({"class": name, "AUC": round(roc_auc, 4)})

        plt.suptitle("ROC Curves — Detection (per class)", fontsize=14)
        plt.tight_layout()

        roc_path = output_dir / "roc_curves_detection.png"
        plt.savefig(roc_path, dpi=120, bbox_inches="tight")
        plt.show()

        df = pd.DataFrame(rows)
        macro = {
            "class": "Macro Average",
            "AUC"  : round(float(np.nanmean(df["AUC"])), 4),
        }
        self._roc_df = pd.concat(
            [df, pd.DataFrame([macro])], ignore_index=True
        )
        self._log.info("ROC AUC:\n%s", self._roc_df.to_string(index=False))
        self._log.info("ROC figure saved → %s", roc_path)
        return self._roc_df

    def plot_training_curves(self, trainer_save_dir: Path) -> None:
        """Plot loss and metric curves from YOLO's results.csv."""
        csv_path = trainer_save_dir / "results.csv"
        if not csv_path.exists():
            self._log.warning("results.csv missing — skipping training curves")
            return

        df = pd.read_csv(csv_path)
        df.columns = [c.strip() for c in df.columns]
        metric_cols = [
            c for c in df.columns
            if any(k in c.lower() for k in ["loss", "map", "precision", "recall"])
        ]
        n_cols = min(4, len(metric_cols))
        n_rows = max(1, (len(metric_cols) + n_cols - 1) // n_cols)

        fig, axs = plt.subplots(n_rows, n_cols, figsize=(5 * n_cols, 4 * n_rows))
        axs      = np.array(axs).flatten()
        x_vals   = df.get("epoch", range(len(df)))

        for i, col in enumerate(metric_cols):
            axs[i].plot(x_vals, df[col])
            axs[i].set_title(col)
            axs[i].set_xlabel("Epoch")

        for j in range(i + 1, len(axs)):
            axs[j].axis("off")

        plt.suptitle("Training Metrics Over Epochs", fontsize=14)
        plt.tight_layout()

        out_path = self.cfg.results_dir / "training_curves.png"
        plt.savefig(out_path, dpi=120, bbox_inches="tight")
        plt.show()
        self._log.info("Training curves saved → %s", out_path)

    def plot_qualitative_predictions(self, n_samples: int = 6) -> None:
        """Overlay YOLO bounding boxes and segmentation masks on random val images."""
        sample_fps   = random.sample(self.val_images, min(n_samples, len(self.val_images)))
        qual_results = self.model.predict(
            source  = [str(fp) for fp in sample_fps],
            imgsz   = self.cfg.train_hp["imgsz"],
            device  = self.cfg.infer_device,
            verbose = False,
            conf    = 0.25,
        )
        cmap = plt.cm.get_cmap("tab10", self.cfg.num_classes)
        fig, axes = plt.subplots(2, 3, figsize=(18, 12))
        axes      = axes.flatten()

        for ax_idx, (fp, res) in enumerate(zip(sample_fps, qual_results)):
            ax     = axes[ax_idx]
            img_np = np.array(Image.open(fp).convert("RGB"))
            ax.imshow(img_np)

            if res.boxes is not None:
                boxes   = res.boxes
                masks_r = res.masks
                for idx in range(len(boxes)):
                    cls_id = int(boxes.cls[idx])
                    conf   = float(boxes.conf[idx])
                    xyxy   = boxes.xyxy[idx].cpu().numpy()
                    color  = cmap(cls_id)

                    ax.add_patch(
                        patches.Rectangle(
                            (xyxy[0], xyxy[1]),
                            xyxy[2] - xyxy[0],
                            xyxy[3] - xyxy[1],
                            linewidth=2,
                            edgecolor=color,
                            facecolor="none",
                        )
                    )
                    ax.text(
                        xyxy[0], xyxy[1] - 4,
                        f"{self.cfg.yolo_to_name.get(cls_id, '?')}  {conf:.2f}",
                        color="white", fontsize=8,
                        bbox=dict(facecolor=color, alpha=0.7, pad=1),
                    )
                    if masks_r is not None and idx < len(masks_r.xy):
                        poly_pts = np.array(masks_r.xy[idx])
                        if len(poly_pts) >= 3:
                            ax.add_patch(
                                patches.Polygon(
                                    poly_pts, closed=True,
                                    facecolor=color, alpha=0.35,
                                )
                            )
            ax.axis("off")
            ax.set_title(fp.stem, fontsize=9)

        plt.suptitle("YOLO Predictions on Validation Set", fontsize=14)
        plt.tight_layout()

        vis_path = self.cfg.results_dir / "qualitative_predictions.png"
        plt.savefig(vis_path, dpi=120, bbox_inches="tight")
        plt.show()
        self._log.info("Qualitative visualisation saved → %s", vis_path)

    def plot_per_class_bar_chart(self) -> None:
        """Grouped bar chart: Precision / Recall / F1 / mAP / mIoU per class."""
        n   = self.cfg.num_classes
        names_short = [self.cfg.yolo_to_name[i][:13] for i in range(n)]

        # Exclude macro-average row from bar data
        det = self._det_df[self._det_df["class"] != "Macro Average"].reset_index(drop=True)
        seg = self._seg_df[self._seg_df["class"] != "Macro Average"].reset_index(drop=True)

        x, w = np.arange(n), 0.15
        fig, ax = plt.subplots(figsize=(max(10, 2 * n), 5))

        ax.bar(x - 2 * w, det["Precision"],      w, label="Precision",  color="steelblue")
        ax.bar(x -     w, det["Recall"],          w, label="Recall",     color="darkorange")
        ax.bar(x,         det["F1"],               w, label="F1",         color="green")
        ax.bar(x +     w, det["mAP@[.5:.95]"],    w, label="mAP@.5:.95", color="red")
        ax.bar(x + 2 * w, seg["mIoU"],             w, label="mIoU",       color="purple")

        ax.set_xticks(x)
        ax.set_xticklabels(names_short, rotation=30, ha="right")
        ax.set_ylim(0, 1.05)
        ax.set_ylabel("Score")
        ax.set_title("Per-Class Metrics — Detection & Segmentation")
        ax.legend(loc="upper right", fontsize=8)
        plt.tight_layout()

        bar_path = self.cfg.results_dir / "per_class_metrics.png"
        plt.savefig(bar_path, dpi=120, bbox_inches="tight")
        plt.show()
        self._log.info("Per-class bar chart saved → %s", bar_path)

    def print_final_summary(self, n_train: int, n_val: int) -> None:
        """
        Emit a consolidated merged summary DataFrame that combines
        detection, segmentation, and AUC metrics in one table,
        matching the final summary block of the original script.
        """
        cfg = self.cfg
        hp  = cfg.train_hp

        self._log.info("=== FINAL METRICS SUMMARY ===")
        self._log.info(
            "GPU(s): %d× %s  |  train=%d  val=%d  epochs=%d  imgsz=%d  batch=%d",
            cfg.num_gpus,
            cfg.gpu_names[0] if cfg.gpu_names else "CPU",
            n_train, n_val,
            hp["epochs"], hp["imgsz"], hp["batch"],
        )

        # Merge detection + segmentation + AUC on the 'class' column
        summary = (
            self._det_df
            .merge(
                self._seg_df[["class", "mIoU", "Dice"]],
                on="class", how="left",
            )
            .merge(
                self._roc_df,
                on="class", how="left",
            )
        )
        self._log.info(
            "Full per-class summary:\n%s", summary.to_string(index=False)
        )

        # Pixel legend
        self._log.info("Semantic mask pixel legend:")
        self._log.info("  pixel 0 = background (unannotated)")
        for ci in range(cfg.num_classes):
            self._log.info(
                "  pixel %d = %s  (YOLO class %d)",
                ci + 1, cfg.yolo_to_name[ci], ci,
            )

        self._log.info("Outputs saved to: %s", cfg.results_dir)

    # ── Private: accumulators ─────────────────────────────────

    def _reset_accumulators(self) -> None:
        n = self.cfg.num_classes
        self.per_class_preds:    dict = {c: [] for c in range(n)}
        self.per_class_gt_cnt:   dict = {c: 0  for c in range(n)}
        self.sem_iou_per_class:  dict = {c: [] for c in range(n)}
        self.sem_dice_per_class: dict = {c: [] for c in range(n)}
        self.y_true_det:         dict = {c: [] for c in range(n)}
        self.y_score_det:        dict = {c: [] for c in range(n)}

    # ── Private: single-image accumulation ────────────────────

    def _accumulate_image(self, img_fp: Path, res) -> None:
        """Process one image result and update all metric accumulators."""
        pil_img = Image.open(img_fp)
        W, H    = pil_img.size
        stem    = img_fp.stem

        gts = self._load_ground_truth(stem, W, H)
        if not gts:
            return

        gt_labels_set = {g["label"] for g in gts}
        gt_by_cls     = defaultdict(list)
        for g in gts:
            gt_by_cls[g["label"]].append(g)
            self.per_class_gt_cnt[g["label"]] += 1

        # ── Parse prediction boxes and masks ─────────────────
        preds = self._extract_preds(res)

        # ── ROC: max-confidence score per class per image ─────
        pred_max_conf = defaultdict(float)
        for p in preds:
            pred_max_conf[p["cls"]] = max(pred_max_conf[p["cls"]], p["conf"])
        for c in range(self.cfg.num_classes):
            self.y_true_det[c].append(1 if c in gt_labels_set else 0)
            self.y_score_det[c].append(pred_max_conf.get(c, 0.0))

        # ── Detection: box IoU matching ───────────────────────
        for p in sorted(preds, key=lambda x: x["conf"], reverse=True):
            c        = p["cls"]
            best_iou = max(
                (self._iou_boxes(p["bbox"], gt["bbox"]) for gt in gt_by_cls.get(c, [])),
                default=0.0,
            )
            self.per_class_preds[c].append((p["conf"], best_iou))

        # ── Segmentation: semantic mask IoU / Dice ────────────
        gt_mask_fp = self.val_mask_dir / f"{stem}.png"
        if not gt_mask_fp.exists():
            return

        gt_sem   = np.array(Image.open(gt_mask_fp))
        pred_sem = np.zeros((H, W), dtype=np.uint8)

        for p in preds:
            if p["conf"] < 0.25 or p["poly"] is None:
                continue
            sem_idx = p["cls"] + 1
            poly_np = np.array(p["poly"], dtype=np.int32).reshape((-1, 1, 2))
            poly_np[..., 0] = np.clip(poly_np[..., 0], 0, W - 1)
            poly_np[..., 1] = np.clip(poly_np[..., 1], 0, H - 1)
            cv2.fillPoly(pred_sem, [poly_np], int(sem_idx))

        for ci in range(self.cfg.num_classes):
            sem_idx  = ci + 1
            gt_bin   = (gt_sem   == sem_idx).astype(np.uint8)
            pred_bin = (pred_sem == sem_idx).astype(np.uint8)
            if gt_bin.sum() == 0:
                continue
            self.sem_iou_per_class[ci].append(self._iou_masks(pred_bin, gt_bin))
            self.sem_dice_per_class[ci].append(self._dice_masks(pred_bin, gt_bin))

    # ── Private: data helpers ─────────────────────────────────

    def _load_ground_truth(self, img_stem: str, img_w: int, img_h: int) -> list:
        """Load YOLO polygon labels and reconstruct bboxes."""
        lbl_fp = self.val_lbl_dir / f"{img_stem}.txt"
        if not lbl_fp.exists():
            return []
        gts = []
        for line in lbl_fp.read_text().splitlines():
            parts = line.strip().split()
            if not parts:
                continue
            cls  = int(parts[0])
            pts  = list(map(float, parts[1:]))
            poly = [
                (pts[i] * img_w, pts[i + 1] * img_h)
                for i in range(0, len(pts) - 1, 2)
            ]
            xs = [p[0] for p in poly]
            ys = [p[1] for p in poly]
            gts.append(
                {
                    "label": cls,
                    "bbox" : [min(xs), min(ys), max(xs), max(ys)],
                    "poly" : poly,
                }
            )
        return gts

    @staticmethod
    def _extract_preds(res) -> list:
        """Unpack YOLO result boxes and mask polygons into a plain list of dicts."""
        preds = []
        if res.boxes is None:
            return preds
        boxes   = res.boxes
        masks_r = res.masks
        for idx in range(len(boxes)):
            poly = (
                masks_r.xy[idx].tolist()
                if masks_r is not None and idx < len(masks_r.xy)
                else None
            )
            preds.append(
                {
                    "cls" : int(boxes.cls[idx]),
                    "conf": float(boxes.conf[idx]),
                    "bbox": boxes.xyxy[idx].cpu().numpy().tolist(),
                    "poly": poly,
                }
            )
        return preds

    # ── Private: metric math (identical to original script) ───

    @staticmethod
    def _iou_boxes(b1: list, b2: list) -> float:
        """Intersection-over-Union for axis-aligned bounding boxes."""
        xi1 = max(b1[0], b2[0]); yi1 = max(b1[1], b2[1])
        xi2 = min(b1[2], b2[2]); yi2 = min(b1[3], b2[3])
        inter = max(0.0, xi2 - xi1) * max(0.0, yi2 - yi1)
        a1 = (b1[2] - b1[0]) * (b1[3] - b1[1])
        a2 = (b2[2] - b2[0]) * (b2[3] - b2[1])
        denom = a1 + a2 - inter
        return inter / denom if denom > 0 else 0.0

    @staticmethod
    def _iou_masks(m1: np.ndarray, m2: np.ndarray) -> float:
        """Binary mask intersection-over-union."""
        inter = np.logical_and(m1, m2).sum()
        union = np.logical_or(m1,  m2).sum()
        return float(inter / union) if union > 0 else 0.0

    @staticmethod
    def _dice_masks(m1: np.ndarray, m2: np.ndarray) -> float:
        """Sørensen–Dice coefficient for binary masks."""
        inter = np.logical_and(m1, m2).sum()
        denom = m1.sum() + m2.sum()
        return float(2 * inter / denom) if denom > 0 else 0.0

    @staticmethod
    def _compute_ap(preds_list: list, n_gt: int, iou_thresh: float = 0.5) -> float:
        """
        11-point interpolated Average Precision.
        Mathematically identical to the original script's ``compute_ap``.
        """
        if n_gt == 0 or not preds_list:
            return 0.0
        sorted_preds = sorted(preds_list, key=lambda x: x[0], reverse=True)
        tp     = np.array([1.0 if iou >= iou_thresh else 0.0 for _, iou in sorted_preds])
        cum_tp = np.cumsum(tp)
        cum_fp = np.cumsum(1.0 - tp)
        prec   = cum_tp / (cum_tp + cum_fp + 1e-9)
        rec    = cum_tp / (n_gt + 1e-9)
        return sum(
            (float(np.max(prec[rec >= r])) if (rec >= r).any() else 0.0) / 11.0
            for r in np.linspace(0.0, 1.0, 11)
        )

    @staticmethod
    def _prf_at_conf(
        preds_list: list,
        n_gt: int,
        conf_thresh: float = 0.25,
        iou_thresh:  float = 0.50,
    ) -> tuple:
        """
        Precision, Recall, F1 at a fixed confidence and IoU threshold.
        Identical formula to the original ``prf_at_conf`` function.
        """
        filtered = [(c, iou) for c, iou in preds_list if c >= conf_thresh]
        tp = sum(1 for _, iou in filtered if iou >= iou_thresh)
        fp = len(filtered) - tp
        fn = max(0, n_gt - tp)
        P  = tp / (tp + fp + 1e-9)
        R  = tp / (tp + fn + 1e-9)
        F1 = 2.0 * P * R / (P + R + 1e-9)
        return P, R, F1


# ════════════════════════════════════════════════════════════
#  ENTRY POINT
# ════════════════════════════════════════════════════════════
def main() -> None:
    log = logging.getLogger(f"{ROOT_LOG.name}.main")

    # ── 1. Configuration ─────────────────────────────────────
    log.info("Initialising DatasetConfig …")
    config = DatasetConfig()
    config.log_summary()

    # ── 2. Build Dataset ─────────────────────────────────────
    log.info("Building YOLO dataset from master JSONs …")
    builder = YoloDataBuilder(config)
    n_train, n_val = builder.build_all()
    builder.compute_class_weights()

    # ── 3. YOLO YAML ─────────────────────────────────────────
    yaml_path = config.generate_yaml()

    # ── 4. Train ─────────────────────────────────────────────
    trainer = YoloTrainer(config)
    trainer.load_pretrained()
    best_model_path = trainer.train(yaml_path)

    # ── 5. Evaluate ──────────────────────────────────────────
    evaluator = YoloEvaluator(config, best_model_path)
    evaluator.run_inference()

    evaluator.compute_detection_metrics()
    evaluator.compute_segmentation_metrics()
    evaluator.compute_roc_auc(config.results_dir)

    evaluator.plot_training_curves(trainer.save_dir)
    evaluator.plot_qualitative_predictions()
    evaluator.plot_per_class_bar_chart()

    evaluator.print_final_summary(n_train, n_val)

    log.info("ALL DONE.  Best model → %s", best_model_path)
    log.info("Outputs   → %s", config.results_dir)


if __name__ == "__main__":
    main()
```
